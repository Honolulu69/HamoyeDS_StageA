```python
#Importing libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
# Loading the dataset
energy = pd.read_csv('energydata_complete.csv')
```

```python
energy.head()
```

```python
#Question 17
from sklearn import metrics
x = energy.iloc[:, 5].values
y = energy.iloc[:, 13].values

x = x.reshape(-1,1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train = x_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
logm = LinearRegression()
logm.fit(x_train, y_train)
y_pred = logm.predict(x_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("The root mean squared error is", round(rmse, 3))
```

```python
# Question 23

# Instantiating ridge model with default alpha value
ridge = Ridge()

ridge.fit(x_train, y_train)

pred1 = ridge.predict(x_test)

ridge_rmse = round(np.sqrt(metrics.mean_squared_error(y_test,pred1)),3)

print(f'The RMSE score for the rigde model with default aplha value when evaluated on the test set is: {ridge_rmse}')
```

```python
#Question 24

# Initiating a lasso regressor with a default alpha value
lasso_reg = Lasso()

lasso_reg.fit(x_train, y_train)

lasso_preds = lasso_reg.predict(x_test)

def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
  weights = pd.Series(model.coef_, feat.columns).sort_values()
  weights_df = pd.DataFrame(weights).reset_index()
  weights_df.columns = ['Features', col_name]
  weights_df[col_name].round(3)
  return weights_df

lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

non_zero_weights = lasso_weights_df[lasso_weights_df['Lasso_weight'] != 0]

# non_zero feature weights 
print(non_zero_weights)
```
